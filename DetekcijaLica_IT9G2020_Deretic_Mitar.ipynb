### **Uopšteno o detekciji lica**

Šta je zapravo detekcija lica? **Detekcija lica se definiše kao postupak izvlačenja** (engl. *extracting*) **lica iz okruženja obuhvaćenog slikom** [2]. Cilj je da sistem uspešno identifikuje određeni region slike kao lice [2]. Bitna razlika u odnosu na prepoznavanje lica je u tome što prepoznavanje **zahteva detekciju samo jednog lica** dok detekcija lica u globalu može podrazumevati **istovremenu detekciju više lica na jednoj slici**, ukoliko ona zaista postoje. Detekcija lica može imati različite primene u odnosu na prepoznavanje lica kao na primer **filteri koji se koriste pri uzimanju fotografije** ili **programi za prebrojavanje ljudi na nekom okupljanju**. Scenario uspešnog detektovanja lica je prikazan na slici ispod.

detekcija_lica_uspesna.jpg

*Slika 2 - Prikaz scenarija uspešne detekcije lica* [5]

Kao što se na slici može primetiti nisu sva lica uspešno detektovana. Razlog je u tome što nisu u potpunosti obuhvaćena slikom a odabrani algoritam u posmatranom slučaju nije uspeo da ih prepozna. Neki od **čestih izazova** koji se javljaju prilikom detekcije lica su već pomenuti u uvodu:

*   **Varijacija poze** - U idealnom slučaju sve osobe obuhvaćene slikom se nalaze u **frontalnoj pozi** (engl. *frontal pose*). Naravno, ovo nije uvek ostvarljivo pogotovo kada slika nije nastala pod kontrolisanim uslovima. Bitni faktori su **poza osobe na slici** i **ugao kamere** [2].
*   **Okluzija** (engl. *occlusion*) **svojstava** - Dešava se u slučaju kada osoba nosi bradu, naočare ili šešir. Može se desiti i u slučaju kada su delovi lica skriveni od kamere [2].
*   **Izraz lica** - Najpovoljniji je **neutralni izraz lica** koji ne menja svojstva koja se posmatraju [2].
*   **Uslovi pod kojim je slika nastala** - Ovde se pre svega misli na **ambijent** i **vrstu kamere** koja je upotrebljena. **Kvalitet slike** može značajno varirati [2].

U zavisnosti od literature, moguće je naići i na termin **lokacija lica** (engl. *face location*). Ovaj termin je **jednostavniji** od termina detekcije lica po tome što podrazumeva prostija rešenja kao što su **pronalaženje površnih svojstava** (na primer lociranje granica lica). Primenjuje se u sistemima gde je potrebno u **realnom vremenu** (engl. *real time*) locirati lice. Primer ovakvog sistema bi bio sistem nadzornih kamera [2].

### **Prilazi detekciji lica**

Postoje različiti pristupi detekciji lica koji se u velikoj meri preklapaju. Nijedan pristup nije naznačen kao najpravilniji ili opšte prihvaćen [2]. U ovom radu biće predstavljene dve klasifikacije pristupa detekciji lica:

1.   **Detekcija koja zavisi od scenarija** [2].
2.   **Detekcija koja je podeljena na kategorije** [2].

**Detekcija koja zavisi od scenarija** razmatra tri scenarija:

*   **Kontrolisano okruženje** (engl. *controlled environment*) - Kontrolisanje osvetljenja, pozadine i ugla pri uzimanju fotografije [2].
*   **Slike u boji** - Detekcija lica se vrši na osnovu boje kože. Ovo može predstavljati problem s obzirom na varijaciju boje ljudske kože kao i raspon od skoro skroz bele do skoro skroz crne. Ipak postoje pokušaji da se pronađe reprezentativna boja kože kako bi se ovaj pristup ostvario [2].
*   **Slike nastale u pokretu** - Ponekad video materijal pruža šansu za detekciju lica. Većina današnjih sistema su u obavezi da poseduju funkcionalnost detekcije lica u pokretu. Veliki je izazov ostvariti kompromis između kvaliteta rezultata i performansi [2].

**Detekcija koja je podeljena na kategorije** podrazumeva četiri kategorije:

*   **Metode bazirane na znanju** (engl. *knowledge-based methods*) - Metode zasnovane na pravilima koje implementiraju naše znanje o ljudskim licima[2].
*   **Metode zasnovane na nemenjajućim svojstvima** (engl. *feature-invariant methods*) - Metode koje pokušavaju da pronađu svojstva koja se ne menjaju na licu bez obzira na ugao i poziciju [2].
*   **Metode zasnovane na podudaranju šablona** (engl. *template matching methods*) - Metode koje vrše poređenje ulaznih slika sa već unapred pripremljenim šablonima lica ili svojstvima [2].  
*   **Metode bazirane na izgledu** (engl. *appearance-based methods*) - Metoda zasnovana na podudaranju šablona kod koje je baza podataka ,,naučena" iz seta slika za trening [2].

### **Koraci algoritma za detekciju lica**

Kao što je u ranijim sekcijama napomenuto, svi algoritmi imaju skoro identične korake, tako da je moguće prikazati određenu vrstu **univerzalnog postupka detekcije lica**. Postupak koji će ovde biti prikazan se pretežno oslanja na **segmentaciju boje** (engl. *color segmentation*), **segmentaciju slike** (engl. *image segmentation*) i **metodu zasnovanu na podudaranju šablona**.

**Segmentacija boje**

Pod segmentacijom boje se podrazumeva detekcija različitih nijansi boja kože na slikama u boji. Ova tehnika je veoma korisna i popularna za detekciju lica. Formati ulaznih slika u boji mogu biti različiti. Najčešći format je ***RGB***. Međutim, *RGB* je osetljiv na promenu osvetljenja slike što bi moglo dovesti do neuspešne detekcije. Iz tog razloga, većina ovakvih tehnika uglavnom koristi komponente boja u **prostoru boja** (engl. *color space*). Najčešći formati su ***HSV*** i ***YIQ***. Postoji mnogo prostora boja, a u ovom primeru biće prikazan ***YCbCr*** prostor boja. Piksel kože u *YCbCr* prostoru boja prikazan je na slikama ispod kao i rezultat segmentacije boje [5].

### **Primer koda za detekciju lica**

Primer koda za detekciju lica na programskoj platformi ***MATLAB*** prikazan je na slici ispod.



```python
function outFaces = faceDetection(img)
%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
% Function 'outFaces' returns the matrix with the information of
% face locations and gender.
%
% outFaces = faceDetection(img)
% img: double formatted image matrix
% coefficients
effect_num=3;
min_face=170;
small_area=15;
imgSize = size(img);
uint8Img = uint8(img);
gray_img=rgb2gray(uint8Img);
% get the image tranformed through YCbCr filter
filtered=ee368YCbCrbin(img,161.9964,-11.1051,22.9265,25.9997,4.3568,3.9479,2);
% black isolated holes rejection
filtered=bwfill(filtered,'holes');
% white isolated holes less than small_area rejection
filtered=bwareaopen(filtered,small_area*10);
% first erosion
filtered = imerode(filtered,ones(2*effect_num));
% edge detection with the Roberts method with sensitivity 0.1
edge_img=edge(gray_img,'roberts',0.1);
% final binary edge image
edge_img=~edge_img;
% integeration of two images, edge + filtered image
filtered=255*(double(filtered) & double(edge_img)); % double
% second erosion
filtered=imerode(filtered,ones(effect_num));
% black isolated holes rejection
filtered=bwfill(filtered,'hole');
% small areas less than the minumum area of face rejection
filtered=bwareaopen(filtered,min_face);
Page 22
% group labeling in the filtered image
[segments, num_segments] = bwlabel(filtered);
% Based on the binary image, squared windows are generated
boxInfo = [];
for i = 1:num_segments,
 [row col] = find(segments == i);
 [ctr, hWdth] = ee368boxInfo(row, col);
 boxInfo = [boxInfo; ctr hWdth];
end
% Overlapping squares are merged
boxInfo = ee368boxMerge(boxInfo, num_segments, imgSize(1));
num_box = length(boxInfo);
% mean squared distance of the boxes wirh respect to others are calculated
boxDist = [];
for i = 1:num_box,
 boxDist = [boxDist; sqrt((sum((boxInfo(:,1) - boxInfo(i,1)).^2) + ...
 sum((boxInfo(:,2) - boxInfo(i,2)).^2))/(num_box-1))];
end

% conversion to 'double' format and 'gray' format
gdOrgImg = double(rgb2gray(uint8Img));
filtered = double(filtered);
outFaces = [];
for k=1:num_box,
 ctr = boxInfo(k, 1:2);
 hWdth = boxInfo(k,3);

 % Based on the box information, images are cut into squares
 testImg = ee368imgCut(gdOrgImg, filtered, imgSize, boxInfo(k,:));

 % normalized with an average brightness
 avgBri = sqrt(sum(sum(testImg.^2))/bwarea(testImg));
 testImg = testImg/avgBri;

 % test images are compared to the eigen images or femaale average images
 corr = ee368imgMatch(testImg, hWdth);
 fCorr = ee368imgMatchFe(testImg, hWdth);
 outFaces = [outFaces; ctr 1 corr/boxDist(k), hWdth, fCorr];
end
% sorting of the correlation values
[Y I] = sort(outFaces(:,4));
outFaces = outFaces(I, :);
% elimination of small correlation values using histogram
B = hist(Y);
if B(1) < 0.5*num_box,
 Page 23
 outFaces = outFaces(B(1)+1:end, :);
end
% results of the correlation with respect to the women's faces
[Fe1, ordFe1] = max(outFaces(:, 6));
[Fe2, ordFe2] = max(outFaces(:, 7));
[Fe3, ordFe3] = max(outFaces(:, 8));
outFaces = outFaces(:, 1:3);
outFaces([ordFe1, ordFe2, ordFe3],3) = 2;
```

Za evaluaciju rezultata upotrebljen je "**koeficijent uspešnosti detekcije lica**". Koeficijent je definisan kao odnos ukupnog broja detektovanih lica i skeniranih slika. Za ovakvu definiciju koeficijenta postoje dva problema:

1.   Postoji mogućnost da je detekcija uspešno izvršena ali da nije tačna.
2.   Na slici se može nalaziti više osoba pa se tako na jednoj slici može detektovati više od jednog lica.
